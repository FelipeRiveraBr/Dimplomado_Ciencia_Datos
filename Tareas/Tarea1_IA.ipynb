## ¿Qué es una medida de dispersión?

Una medida de dispersión es una herramienta que nos ayuda a cuantificar la distancia relativa entre los datos y el centro de distribución. De alguna manera nos da información sobre qué tan contraída o estirada es la distribución de los datos

### Ejemplo

Entre las medidas de dispersión más conocidas se encuentra la varianza que se simboliza con $\sigma^{2}$ y se define como: 

 $\sigma^{2}_{X}=E[(X-\mu)^{2}]$ 

Donde $X$ es una variable aleatoria y $\mu$ es su media
